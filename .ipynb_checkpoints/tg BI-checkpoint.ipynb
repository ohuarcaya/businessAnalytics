__Filtraremos la data con NaN o NULL__

```python
diasxmes = 30
# cambiaremos las fechas por sus diferencias en días
df.dropna(subset=['fechaVencimiento'], inplace=True) # quitamos data sin fecha de vencimiento
df['tiempoLimite'] = (df.fechaVencimiento - df.fechaApertura).dt.days/diasxmes
df['tiempoPagando'] = (df.fechaUltimoPago - df.fechaApertura).dt.days/diasxmes
df.drop(['fechaApertura'], axis=1, inplace=True)
df.drop(['fechaUltimoPago'], axis=1, inplace=True)
df.drop(['fechaVencimiento'], axis=1, inplace=True)
df['tiempoPagando'].fillna(0, inplace=True) # quienes no hicieron ningún pago
# Definimos quienes son deudores aún 
df['esDeudor'] = np.int32(df.dSalPag.isnull())#1:deudor, 0:no deudor
df.drop(['dSalPag'], axis=1, inplace=True)
df.isnull().sum() # buscar campos sin datos
```
